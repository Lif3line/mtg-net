```python
import pandas as pd
import numpy as np
import json

file_path = "data/AllSets.json"

with open(file_path) as data_file:  
    data = json.load(data_file)
```

```python
i = 0
for key in data:
    print ("key: {}".format(key))
    i += 1
    
print("\nSets: {}".format(i))
```
